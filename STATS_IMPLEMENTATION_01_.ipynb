```python
#t distribstion t dist with df 1

t = np.random.standard_t(df = 1, size = 10000)
t = t[np.where(abs(t)<10)]
plt.hist(t , bins = 100 )
plt.show()
```

```python
#t distribstion t dist with df 10 >> tail is thicker/larger
t = np.random.standard_t(df = 10, size = 10000)
t = t[np.where(abs(t)<10)]
plt.hist(t , bins = 100 )
plt.show()
```

```python
#F distribution
#dfnum is df for numerator, dfden is df for den
f = np.random.f(dfnum = 1, dfden=1, size=10000)
f = f[np.where(abs(f) < 10)]
plt.hist(f, bins = 1000)
plt.show()
```

```python
f = np.random.f(dfnum = 3, dfden=10, size=10000)
f = f[np.where(abs(f) < 10)]
plt.hist(f, bins = 1000)
plt.show()
```
